# **7. What is the standard deviation of pressure in the data**

# **9. Find all the instances when 'Snow' was recorded**

```python
data[data['Weather Condition'].str.contains('Snow')]
```

# **9. Find all the instances when wind speed is above 24 and visibility as 25**

```python
data[(data['Wind Speed_km/h']>24) & (data['Visibility_km']==25)]
```

# **10. What is the mean value of each column against each 'Weather condition'**

```python
data.groupby(['Weather Condition']).mean(numeric_only=True)
```

# **11. What is the minimum and maximum value of each column against each weather condition**

```python
data.groupby(['Weather Condition']).max(numeric_only=True)
```

```python
data.groupby(['Weather Condition']).min(numeric_only=True)
```

# **13. Show all the records where weather condition is Fog**

# **14. Find all instances where weather is clear or visibility is above 40**

```python
data[(data['Weather Condition']=='Clear') | (data['Visibility_km']>40)]
```

# **15. Find all the instances where weather is clear and relative humidity is greater than 50 or visibility is above 40**

```python
data[(data['Weather Condition']=='Clear') & (data['Rel Hum_%']>50) | (data['Visibility_km']>40)]
```
